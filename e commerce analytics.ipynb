## Day when customers bought most items. for every item 

```python
df_ordered=pd.merge(df2,df3, how='left', on='order_id')
```

```python
df_ordered
```

```python
df_ordered['order_purchase_timestamp'] = pd.to_datetime(df_ordered['order_purchase_timestamp'], errors='coerce')
```

```python
df_ordered['day_month']=df_ordered['order_purchase_timestamp'].dt.day_name()
```

```python
df_ordered.product_id.nunique()
```

```python
df_ordered=df_ordered.groupby(['product_id','day_month'], as_index=False).agg({'order_id':'count'})
df_ordered
```

```python
df_ordered=df_ordered.pivot(index='product_id', columns='day_month', values='order_id').idxmax(axis=1).to_frame().reset_index()
```

```python
df_ordered.rename(columns={'product_id':'item', 0:'most_purchases_per_day'})
```
